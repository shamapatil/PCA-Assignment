```python
### Converting data to numpy array

winedata_array = wine2.values
winedata_array
```

```python
# Normalizing the numerical data 
from sklearn.preprocessing import scale
wine_norm=scale(winedata_array)
wine_norm
```

**PCA Implementation**

```python
### Applying PCA fit Transform to dataset

pca = PCA(n_components=13)
wine_pca = pca.fit_transform(wine_norm)
wine_pca
```

```python
### PCA components matrix or covariance matrix
pca.components_
```

```python
# The amount of variance that each PCA has :-

var = pca.explained_variance_ratio_
var 
```

```python
f,ax = plt.subplots(figsize=(10,8))
plt.bar(range(1, len(var)+1),var)
plt.xlabel('Number of Components')
plt.ylabel('variance (%)')
plt.title('Explained variance by each component')
```

```python
### Cumulative variance

var1 = np.cumsum(np.round(var,4)*100)
var1
```

```python
## Variance plot for PCA components obtained
f,ax = plt.subplots(figsize=(6,6))
plt.plot(var1, color='magenta');
```

Checking with other Clustering Algorithms

Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
## Creating dendograms

plt.figure(figsize=(10,8))
dendrogram = sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
# Creating clusters 
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y = pd.DataFrame(hclusters.fit_predict(wine_norm), columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
## adding clusters to data set

winedata3 = wine.copy()
winedata3['clustersid']=hclusters.labels_
winedata3
```

K-Means Clustering

```python
from sklearn.cluster import KMeans
## WCSS - within-cluster sum of squares criterion

wcss=[]
for i in range(1,6):
    kmeans = KMeans(n_clusters=i, random_state=2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
## Plot Elbow graph
plt.figure(figsize=(8,6))
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

Building cluster algorithm using k=3

```python
# Cluster algorithm using k=3

clusters3 = KMeans(3,random_state=30).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
# Assign clusters to data set

winedata4 =wine.copy()
winedata4['clusters3id']=clusters3.labels_
winedata4
```
